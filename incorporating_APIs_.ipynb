```python
from google.colab import userdata

import os

from langchain_tavily import TavilySearch

#setting up my API key as an environment variable

os.environ["TAVILY_API_KEY"] = userdata.get('TAVILY_API_KEY')
```

```python

# Create the search tool

search_tool = TavilySearch(

    max_results=5,

    search_depth="basic",

    include_answer=True,

    include_raw_content=False

)


# Test the web search tool with a question about a recent event

result = search_tool.invoke({"query": "Who won album of the year at the \most recent Grammys?"})

print(result['answer'])
```
